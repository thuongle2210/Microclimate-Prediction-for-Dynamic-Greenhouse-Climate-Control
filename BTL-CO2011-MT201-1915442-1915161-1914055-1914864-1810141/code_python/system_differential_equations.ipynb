```python
import matplotlib.pyplot as plt
import numpy as np           
import math
```

```python
h=0.1
t=np.arange(0,0.6,0.1)
x1=[math.exp(4*x)+3*math.exp(-x) for x in t]
x2=[-math.exp(4*x)+2*math.exp(-x) for x in t]
#plt.show()
```

Solve System of Differential equations using Euler's method

```python
Predict_x1=[4]
Predict_x2=[1]
k=0
for i in range(0,len(t)-1):
    x1_derivative=Predict_x1[k]-3*Predict_x2[k]
    x2_derivative=-2*Predict_x1[k]+2*Predict_x2[k]
    #print(x1_derivative)
    #print(x2_derivative)
    Predict_x1.append(Predict_x1[k]+h*x1_derivative)
    Predict_x2.append(Predict_x2[k]+h*x2_derivative)
    k=k+1
plt.plot(t,Predict_x1,label='Predict x1')
plt.plot(t,x1,label='Actual x1')
plt.plot(t,Predict_x2,label='Predict x2')
plt.plot(t,x2,label='Actual x2')
plt.xlabel("t")
plt.legend(loc='best')

print(Predict_x1)
print(Predict_x2)
```

Using Runge-kutta

```python
def x1_derivative(x1,x2):
    return x1-3*x2
def x2_derivative(x1,x2):
    return -2*x1+2*x2
Predict_x1=[4]
Predict_x2=[1]
for i in range(0,len(t)-1):
    k11=h*x1_derivative(Predict_x1[i],Predict_x2[i])
    k12=h*x2_derivative(Predict_x1[i],Predict_x2[i])
    #print("k11=",k11)
    #print("k12=",k12)
    k21=h*x1_derivative(Predict_x1[i]+k11/2,Predict_x2[i]+k12/2)
    k22=h*x2_derivative(Predict_x1[i]+k11/2,Predict_x2[i]+k12/2)
    #print("k21=",k21)
    #print("k22=",k22)
    k31=h*x1_derivative(Predict_x1[i]+k21/2,Predict_x2[i]+k22/2)
    k32=h*x2_derivative(Predict_x1[i]+k21/2,Predict_x2[i]+k22/2)
    #print("k31=",k31)
    #print("k32=",k32)
    k41=h*x1_derivative(Predict_x1[i]+k31,Predict_x2[i]+k32)
    k42=h*x2_derivative(Predict_x1[i]+k31,Predict_x2[i]+k32)
    #print("k41=",k41)
    #print("k42=",k42)
    #print("Delta1=",(k11+2*k21+2*k31+k41)/6)
    #print("Delta2=",(k12+2*k22+2*k32+k42)/6)
    Predict_x1.append(Predict_x1[i]+(k11+2*k21+2*k31+k41)/6)
    Predict_x2.append(Predict_x2[i]+(k12+2*k22+2*k32+k42)/6)
    #print("Predict1=",Predict_x1)
    #print("Predict2=",Predict_x2)
plt.plot(t,Predict_x1,label='Predict x1')
plt.plot(t,x1,label='Actual x1')
plt.plot(t,Predict_x2,label='Predict x2')
plt.plot(t,x2,label='Actual x2')
plt.xlabel("t")
plt.legend(loc='best')
plt.show()
```
